```python
import os
import numpy as np
import pandas as pd
```

```python
raw_topic_matrix = np.random.exponential(scale=5, size=(30,15))
```

```python
raw_topic_matrix
```

```python
normalized_topic_matrix = raw_topic_matrix/np.asarray([np.sum(raw_topic_matrix,axis=1),]*15).transpose()
```

```python
normalized_topic_matrix
```

```python
raw_query_vector = np.random.exponential(scale=5, size=15)
```

```python
normalized_query_vector = [float(i)/sum(raw_query_vector) for i in raw_query_vector]
```

```python
normalized_query_vector
```

```python
type(normalized_query_vector)
```

```python
normalized_topic_matrix = pd.DataFrame(normalized_topic_matrix)
```

```python
normalized_topic_matrix.index.values
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(normalized_topic_matrix, normalized_topic_matrix.index.values)
labels = knn.predict(np.array([normalized_query_vector]))
```

```python
print(labels)
```

```python
from sklearn.neighbors import KDTree
tree = KDTree(normalized_topic_matrix)
dist, ind = tree.query(np.array([normalized_query_vector]), k=5)
print(ind)  # indices of 5 closest neighbors
print(dist)  # distances to 5 closest neighbors
```
